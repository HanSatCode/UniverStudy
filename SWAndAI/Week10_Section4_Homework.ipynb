```python
import numpy as np
import matplotlib.pyplot as plt
data = {'X' : np.random.rand(25), 'Y' : np.random.rand(25), 'S' : np.random.randint(100, 1000, 25), 'C' : np.random.randint(1, 5, 25)}
plt.scatter(data['X'], data['Y'], s=data['S'], c = data['C'])
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
